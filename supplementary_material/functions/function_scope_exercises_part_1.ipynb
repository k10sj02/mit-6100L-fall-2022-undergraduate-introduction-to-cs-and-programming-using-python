### üîç **Exercise A: Local Modification**

```python
def f(x):
    x = x + 1
    print('in f(x): x =', x)
    return x

x = 3
z = f(x)
print('x =', x)
print('z =', z)
```

### üîç **Exercise B: Mutating vs Rebinding**

```python
def f(a):
    a.append(4)
    print('in f(a): a =', a)

a = [1, 2, 3]
f(a)
print('a =', a)
```

### üîç **Exercise C: No Return, No Change**

```python
def f(x):
    x = x * 2
    print('in f(x): x =', x)

x = 5
f(x)
print('x =', x)
```

### üîç **Exercise D: Return, Reassign**

```python
def f(x):
    x = x ** 2
    print('in f(x): x =', x)
    return x

x = 4
x = f(x)
print('x =', x)
```

---

### üîç **Exercise E: Global Rebinding**

```python
def f():
    global x
    x = x + 2
    print('in f(): x =', x)

x = 1
f()
print('x =', x)
```

### üîç **Exercise F: Immutable Passed In**

```python
def double(num):
    num *= 2
    print("Inside double:", num)

x = 10
double(x)
print("After function call:", x)
```

### üîç **Exercise G: Return but Don‚Äôt Assign**

```python
def subtract(y):
    y = y - 5
    print("Inside subtract:", y)
    return y

x = 20
subtract(x)
print("Outside subtract:", x)
```

````python
# 1. Maximum of Three Numbers  
# Write a Python function to find the maximum of three numbers.  

def three_num_max(n1, n2, n3):
  """
  Return the largest of three numbers
  """
  largest = max(n1, n2, n3)
  return largest

three_num_max(1, 78, 4)
````

````python
# Write a Python function to sum all the numbers in a list.  
# **Sample List**: `(8, 2, 3, 0, 7)`  
# **Expected Output**: `20`  

def sum_list_num(my_list):
  """
  Return the sum of all numbers in a list
  """
  result = sum(my_list)
  print(result)

sum_list_num([8, 2, 3, 0, 7])
````

````python
# Write a Python function to multiply all the numbers in a list.  
# **Sample List**: `(8, 2, 3, -1, 7)`  
# **Expected Output**: `-336`  

def mult_list_num(my_list):
  """ 
  Return the product of all numbers in a list
  """
  index = 0 
  result = 1
  while index < len(my_list):
    result *= (my_list[index])
    index += 1
  return result
  
mult_list_num([1,2,3])
````

2. Which of the following expressions result in `SyntaxError`s?

   - `6 * ———--8`
   - `8 = people`
   - `((((4 ** 3))))`
   - `(-(-(-(-5))))`
   - `4 += 7 / 2`

```python
# 2. Which of the following expressions result in SyntaxErrors?

# ❌ `6 * ———--8`
# This will cause a SyntaxError because it uses em dashes (—) instead of regular minus signs (-).
# Python does not recognize em dashes as valid syntax.

# ❌ `8 = people`
# This will cause a SyntaxError because you cannot assign a value to a literal like 8.
# The left side of an assignment must be a variable, not a number.

# ✅ `((((4 ** 3))))`
# This is valid syntax.
# Python allows nested parentheses like this. It just evaluates 4 ** 3 = 64.

# ✅ `(-(-(-(-5))))`
# This is also valid.
# It’s just multiple negations. Final result is -5.

# ❌ `4 += 7 / 2`
# This will cause a SyntaxError.
# You can't use += with a literal like 4. The left-hand side must be a variable.

# ✅ Final Answer: The following expressions result in SyntaxErrors:
#  - `6 * ———--8`
#  - `8 = people`
#  - `4 += 7 / 2`
```

```python
# For each of the following expressions, what value will the expression give?
# Verified in Python:

# 9 - 3
# ➜ 6
# Subtraction — basic arithmetic.

# 8 * 2.5
# ➜ 20.0
# Multiplication with a float returns a float.

# 9 / 2
# ➜ 4.5
# Division always returns a float in Python 3.

# 9 / -2
# ➜ -4.5
# Division with a negative divisor.

# 9 // -2
# ➜ -5
# Floor division — rounds **down** to the nearest whole number (toward -infinity).

# 9 % 2
# ➜ 1
# Remainder of 9 divided by 2.

# 9.0 % 2
# ➜ 1.0
# Same as above, but result is a float because 9.0 is a float.

# 9 % 2.0
# ➜ 1.0
# Again, float result because of the float divisor.

# 9 % -2
# ➜ -1
# Remainder takes the sign of the divisor (-2), so result is -1.

# -9 % 2
# ➜ 1
# Remainder takes the sign of the **divisor** (2), so result is positive.

# 9 / -2.0
# ➜ -4.5
# Regular division with float result.

# 4 + 3 * 5
# ➜ 19
# Multiplication happens before addition → 3*5 = 15 → 4 + 15 = 19.

# (4 + 3) * 5
# ➜ 35
# Parentheses come first → 4 + 3 = 7 → 7 * 5 = 35.
```
